```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
flights_ = pd.read_csv("E:\Data Analyst Coding/Machine Learning Project Internship/Travel_MLOps_Project\Data/flights.csv")
flights_
```

```python
flights_["date"] = pd.to_datetime(flights_["date"])
flights_["date"]
```

```python
flights_["TravelMonth"] = flights_["date"].dt.month

flights_["TravelDay"] = flights_["date"].dt.day

flights_["TravelWeekday"] = flights_["date"].dt.weekday

flights_["Weekend"] = (flights_["TravelWeekday"] >= 5).astype(int)
```

```python
flights_.drop("date", axis = 1, inplace = True)
```

```python
from_encoder = LabelEncoder()
to_encoder = LabelEncoder()
flight_encoder = LabelEncoder()
agency_encoder = LabelEncoder()

flights_["from"] = from_encoder.fit_transform(flights_["from"])

flights_["to"] = to_encoder.fit_transform(flights_["to"])

flights_["flightType"] = flight_encoder.fit_transform(flights_["flightType"])

flights_["agency"] = agency_encoder.fit_transform(flights_["agency"])
```

```python
scaler = StandardScaler()

flights_[["time", "distance"]] = scaler.fit_transform(
    flights_[["time", "distance"]]
)
```

```python
X = flights_.drop("price", axis=1)

y = flights_["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators = 100, random_state = 42)

model.fit(X_train, y_train)
```
